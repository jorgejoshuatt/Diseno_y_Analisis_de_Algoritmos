```
def ejemplo1(n):
  c=n+1
  d=c*n
  e=n*n
  total=c+e-d
  print(f"total={total}")
ejemplo1(99999)
#4
```

```
def ejemplo2( n ):
    contador = 0
    for i in range( n ) :
        for j in range( n ) :
            contador += 1
    return contador

ejemplo2( 20 )
#1+n^2
```

```
def ejemplo4( n ):
    x = 3 * 3.1416 + n
    y = x + 3 * 3 - n
    z = x + y
    return z
ejemplo4(9)
#3
```

```
def ejemplo5( x ):
    n = 10
    for j in range( 0 , x , 1 ):
        n = j + n
    return n
ejemplo5(10)
#1+x
```

```
def ejemplo6( n ):
    data=[[[1 for x in range(n)] for x in range(n)]
          for x in range(n)]
    suma = 0
    for d in range(n):
        for r in range(n):
            for c in range(n):
                suma += data[d][r][c]
    return suma
print(ejemplo6( 5 ))
#n^3+2
```

```
from time import time
def ejemplo6( n ):
    start_time = time()
    data=[[[1 for x in range(n)] for x in range(n)] 
          for x in range(n)]
    suma = 0
    for d in range(n):
        for r in range(n):
            for c in range(n):
                suma += data[d][r][c]
    elapsed_time = time() - start_time
    print("Tiempo transcurrido: %0.10f segundos." % elapsed_time)
    return suma

for entrada in range(100,600,100):
  ejemplo6(entrada)
```

```
def ejemplo7( n ):
    count = 0
    for i in range( n ) :
        for j in range( 25 ) :
            for k in range( n ):
                count += 1
    return count

ejemplo7(3)

#1+25n^2
```

```
def ejemplo7_2( n ):
    count = 0
    for i in range( n ) :
        for j in range( 25 ) :
            for k in range( n ):
                count += 1
            for k in range( n ):
                count += 1
    return count 
    
# T(n)= 1 + 50n^2
# O(n)= n^2

ejemplo7_2(3)
```

```
numeros=[1,2,3,4,5,6]
def ejemplo8( numeros ): # numeros es una lista (arreglo en c)
    total = 0
    for index in range(len(numeros)):
        total = numeros[index]
    return total
ejemplo8(numeros)

#T(n)= 1+n
#O(n)=n
```

```
def ejemplo10( n ):
    count = 0
    for i in range( n ) :
        for j in range( i+1 ) :
            count += 1
    return count

print(ejemplo10( 6 ))
#T(n)=1+(n^2+n)/2
#O(n)=n^2
```
