```
class Stack:
  def __init__(self):
    self.__datos=[] #En java se puede meter arraylist

  def is_empty(self):
    return len(self.__datos)==0

  def get_top(self):
    return self.__datos[len(self.__datos)-1]

  def pop(self):
    return self.__datos.pop()

  def push(self,valor):
    self.__datos.append(valor)

  def get_length(self):
    return len(self.__datos)

  def to_string(self):
    print("|---------|")
    for i in self.__datos[-1::-1]:
      print(i)
    print("|---------|")
```

```
nombre_archivo = "Busqueda_binaria.c"
with open(nombre_archivo, "r") as archivo:
    contenido = archivo.read()
    print("El contenido es:\n\n", contenido)
    i= 0   
    pila= Stack()
    for i in range(0,len(contenido)):
      simbolo = contenido[i]
      if simbolo=="(" or simbolo=="[" or simbolo=="{":
        pila.push("@")
      elif simbolo==")" or simbolo=="]" or simbolo=="}":
        pila.pop()
    if pila.is_empty():
      print("\n\n\nLA CADENA ESTÁ BALANCEADA")
    else:
      print("\n\n\nLA CADENA NO ESTÁ BALANCEADA")

    
```
